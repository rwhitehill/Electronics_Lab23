```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
params = {}
params['text.latex.preamble'] = r'\usepackage{amsmath}'
params['text.usetex'] = True
params['font.family'] = 'sans-serif'
params['font.sans-serif'] = ['Helvetica']
plt.rcParams.update(params)
```

```python
# with plt.xkcd():
fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(7,5))

x = np.linspace(-4,2)
y =  (x+1)**2 + 0.5

ax.spines[['top','right','left','bottom']].set_visible(False)

ax.axhline(y=0,color='k',lw=5)
ax.axvline(x=0,color='k',lw=5)

ax.plot(x,y,alpha=0.7,lw=3)
ax.axhline(y=7,color='r',alpha=0.7,lw=3)

ax.text(x=2,y=0.2,s=r'$\Delta g$',va='bottom',ha='right',size=30)
ax.text(x=-3.5,y=7.2,s=r'$A_{LL}^{\rm jet}$',va='bottom',ha='left',size=30)

ax.tick_params(which='major',axis='both',labelsize=0)
ax.tick_params(left=False,bottom=False)

plt.show()
fig.savefig('delta-g_fit-sketch.pdf',bbox_inches='tight')
```
