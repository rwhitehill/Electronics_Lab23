```python
from scipy.integrate import quad
import matplotlib.pyplot as plt
import numpy as np
```

```python
params = {}
params['text.latex.preamble'] = r'\usepackage{amsmath}'
params['text.usetex'] = True
params['font.family'] = 'sans-serif'
params['font.sans-serif'] = ['Helvetica']
plt.rcParams.update(params)
```

```python
def H(t):
    if t >= 0:
        return 1
    else:
        return 0

def V(t):
    # return V0*np.sin(2*np.pi*f*t)
    periods = int(t/T)
    if periods % 2 == 0:
        return V0
    else:
        return 0
```

```python
def integrand(t):
    return np.exp(t/R/C)*V(t)

def Q(t):
    return np.exp(-t/R/C)*quad(integrand,0,t,limit=1000)[0]/R

def V_C(t):
    return Q(t)/C

def V_R(t):
    return (V(t)-V_C(t))
```

```python
R  = 6e3
C  = 3e-9
V0 = 5.0/2.0

f = 2503#1/(2*np.pi*R*C)
T = 1/f
```

```python
R*C
```

```python
t = np.linspace(0,2*T,1000)
V_ = np.array([V(_) for _ in t])
VC = np.array([V_C(_) for _ in t])

reduced_t = t*1e3

fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(7,5))

ax.plot(reduced_t,V_,'k-.',label=r'$\rm function \, generator$')
ax.plot(reduced_t,VC,label=r'$\rm capacitor$')

ax.set_ylim(bottom=-1)
ax.legend(loc='lower center',fontsize=20,frameon=False)
ax.set_xlabel(r'$t~[{\rm ms}]$',size=20)
ax.set_ylabel(r'$V~[{\rm volt}]$',size=20)
ax.tick_params(which='major',axis='both',direction='in',labelsize=20)

plt.show()
fig.savefig('capacitor_charge-discharge.pdf',bbox_inches='tight')
```

```python
R   = 6e3
C   = 3e-9
tau = R*C
V0  = 5.0/2.0

F = np.array([[10,20,50],[100,200,500],[1000,2000,5000],[10000,20000,50000]])#10000#/(2*np.pi*R*C)
T = 1/F
```

```python
V   = lambda t,f: V0*np.sin(2.0*np.pi*f*t)
V_C = lambda t,f: V0*(np.sin(2.0*np.pi*f*t)-2.0*np.pi*f*tau*np.cos(2*np.pi*f*t)+2.0*np.pi*f*tau*np.exp(-t/tau))/(4.0*np.pi**2.0*f**2.0*tau**2.0+1.0)
V_R = lambda t,f: V(t,f) - V_C(t,f)
```

```python
nrows,ncols = np.shape(F)
fig,ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=(7*ncols,5*nrows))

for i in range(nrows):
    for j in range(ncols):
        t = np.linspace(0,3*T[i][j],100)
        V_ = np.array([V(_,F[i][j])   for _ in t])
        VC = np.array([V_C(_,F[i][j]) for _ in t])
        VR = np.array([V_R(_,F[i][j]) for _ in t])
        
        ax[i][j].plot(t,V_,'k-.')#,label=r'$\rm source$'  if i==0 and j == 0 else '')
        ax[i][j].plot(t,VC,'b',alpha=0.5)#,label=r'$\rm capacitor$' if i==0 and j == 0 else '')
        ax[i][j].plot(t,VR,'r',alpha=0.5)#,label=r'$\rm resistor$'  if i==0 and j == 0 else '')
        ax[i][j].tick_params(which='major',axis='both',direction='in',labelsize=20)
        ax[i][j].set_ylim(top=3.5)
        ax[i][j].text(x=0.95,y=0.95,s=r'\boldmath $f = %s\,{\rm Hz}$'%F[i][j],size=20,va='top',ha='right',transform=ax[i][j].transAxes)
        
# ax[0][0].legend(fontsize=20,loc=1,frameon=False)
for i in range(nrows):
    ax[i][0].set_ylabel(r'$V~[{\rm volt}]$',size=20)
for j in range(ncols):
    ax[-1][j].set_xlabel(r'$t~[{\rm s}]$',size=20)
    
plt.tight_layout()
plt.show()
fig.savefig('high_low-pass.pdf')
```
